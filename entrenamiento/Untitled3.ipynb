Realizar un análisis estadístico descriptivo completo de un conjunto de datos de ejemplo, separando las columnas numéricas y categóricas para calcular sus respectivas estadísticas descriptivas y la frecuencia de valores para las columnas categóricas.

## Cargar un Conjunto de Datos de Ejemplo

### Subtask:
Cargar un conjunto de datos de ejemplo de una biblioteca de Python (por ejemplo, scikit-learn o seaborn) para el análisis.


**Reasoning**:
The subtask requires loading the Iris dataset from scikit-learn and converting it into a pandas DataFrame, including feature names and the target column. This will be accomplished by importing necessary libraries and functions, loading the data, and then constructing the DataFrame as specified.



```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Create a pandas DataFrame from the data, using feature names for columns
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add the 'target' column to the DataFrame
iris_df['target'] = iris.target

# Display the first few rows of the DataFrame
print("Iris dataset loaded successfully and converted to DataFrame.")
iris_df.head()
```

## Separar Columnas Numéricas y Categóricas

### Subtask:
Identificar y separar las columnas del DataFrame en dos grupos: numéricas y categóricas.


**Reasoning**:
The subtask requires identifying and separating numerical and categorical columns from the `iris_df` DataFrame. I will use `select_dtypes` to get numerical columns and explicitly define 'target' as categorical, then store their names in separate lists.



```python
numerical_cols = iris_df.select_dtypes(include=np.number).columns.tolist()
categorical_cols = ['target']

# Remove 'target' from numerical_cols if it was included by select_dtypes
# (though in this specific Iris case, 'target' is numerical, so it needs to be explicitly moved to categorical)
if 'target' in numerical_cols:
    numerical_cols.remove('target')

print("Columnas Numéricas:", numerical_cols)
print("Columnas Categóricas:", categorical_cols)
```

## Calcular Estadísticas Descriptivas Numéricas

### Subtask:
Calcular estadísticas descriptivas para las columnas numéricas (media, desviación estándar, mínimo, máximo, cuartiles, etc.) y presentarlas en una tabla.


**Reasoning**:
The subtask requires calculating descriptive statistics for the numerical columns identified previously. Using the `describe()` method on the selected numerical columns of the DataFrame will provide the requested statistics (mean, standard deviation, min, max, quartiles).



```python
numerical_stats = iris_df[numerical_cols].describe()
print("Estadísticas Descriptivas para Columnas Numéricas:")
print(numerical_stats)
```

## Calcular Estadísticas Descriptivas Categóricas

### Subtask:
Calcular estadísticas descriptivas para las columnas categóricas (conteo, únicos, moda, frecuencia de la moda, etc.) y presentarlas en una segunda tabla.


**Reasoning**:
The subtask requires calculating descriptive statistics for the categorical columns. I will apply the `describe()` method to the `categorical_cols` of the `iris_df`.



```python
categorical_stats = iris_df[categorical_cols].describe()
print("Estadísticas Descriptivas para Columnas Categóricas:")
print(categorical_stats)
```

**Reasoning**:
The previous `describe()` method, when applied to the 'target' column which contains numerical values, treated it as a numerical column. To accurately fulfill the subtask's requirement for categorical statistics (count, unique, mode, mode frequency), I need to explicitly convert the 'target' column to a categorical data type before applying `describe()`.



```python
categorical_stats_corrected = iris_df[categorical_cols].astype('category').describe()
print("Estadísticas Descriptivas para Columnas Categóricas (Corregidas):")
print(categorical_stats_corrected)
```

## Calcular Frecuencia de Valores Categóricos

### Subtask:
Calcular la frecuencia de valores (value counts) para cada columna categórica y presentar estos resultados en una tercera tabla o un conjunto de tablas para cada columna.


**Reasoning**:
To calculate and display the frequency of values for each categorical column, I will iterate through the `categorical_cols` list, apply `value_counts()` to each corresponding column in `iris_df`, and print the results with a clear header for each column.



```python
print("\nFrecuencia de Valores para Columnas Categóricas:")
for col in categorical_cols:
    print(f"\nFrecuencia de Valores para '{col}':")
    print(iris_df[col].value_counts())
```
