```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
data = load_iris()
y = data.target
x = data.data
pca = PCA(n_components=2)
x_reducer=pca.fit_transform(x)
print(x_reducer[:5])
```

```python
from sys import platlibdir
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris
data = load_iris()
df_pca= pd.DataFrame(x_reducer,columns=['pca1','pca2'])
df_pca['Species']=data.target
df_pca['Species Name']= df_pca['Species'].apply(lambda x:data.target_names[x])
plt.figure(figsize=(10,6))
sns.scatterplot(
    x='pca1',
    y='pca2',
    hue='Species Name',
    data=df_pca,
    palette='viridis',
    s=100,
    alpha=0.6,
)
plt.title("2 component")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()
```

```python
dfa_corr = df_pca.drop(columns=['Species', 'Species Name'])
corr_matrix = dfa_corr.corr()
print(corr_matrix)
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```
