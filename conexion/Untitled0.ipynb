## Analyze Referee Card Counts

### Subtask:
Group the data by 'Referee' and calculate the total yellow and red cards issued by each referee (summing HY, AY, HR, AR). Visualize the total card counts per referee using a bar chart with appropriate legends.


**Reasoning**:
To analyze the card counts per referee, I will first group the DataFrame by 'Referee' and sum the yellow and red card columns. Then, I will calculate the 'TotalCards' for each referee, sort them, and finally visualize the results using a bar plot to identify referees who issued the most cards, as requested.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select the card columns
card_cols = ['HY', 'AY', 'HR', 'AR']

# Group by Referee and sum the card columns
referee_card_counts = df.groupby('Referee')[card_cols].sum().reset_index()

# Calculate TotalCards
referee_card_counts['TotalCards'] = referee_card_counts[card_cols].sum(axis=1)

# Sort by TotalCards in descending order
referee_card_counts = referee_card_counts.sort_values(by='TotalCards', ascending=False)

# Create the bar plot
plt.figure(figsize=(12, 7))
sns.barplot(x='Referee', y='TotalCards', data=referee_card_counts, palette='viridis')
plt.title('Total Cards Issued per Referee')
plt.xlabel('Referee')
plt.ylabel('Total Cards')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

**Reasoning**:
The previous code block produced a `FutureWarning` from `seaborn` regarding the use of `palette` without `hue`. To resolve this and align with future `seaborn` versions, I will explicitly set `hue='Referee'` and `legend=False` in the `sns.barplot` function, as suggested by the warning.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select the card columns
card_cols = ['HY', 'AY', 'HR', 'AR']

# Group by Referee and sum the card columns
referee_card_counts = df.groupby('Referee')[card_cols].sum().reset_index()

# Calculate TotalCards
referee_card_counts['TotalCards'] = referee_card_counts[card_cols].sum(axis=1)

# Sort by TotalCards in descending order
referee_card_counts = referee_card_counts.sort_values(by='TotalCards', ascending=False)

# Create the bar plot
plt.figure(figsize=(12, 7))
sns.barplot(x='Referee', y='TotalCards', data=referee_card_counts, palette='viridis', hue='Referee', legend=False)
plt.title('Total Cards Issued per Referee')
plt.xlabel('Referee')
plt.ylabel('Total Cards')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Analyze Team Home/Away Goal Performance

### Subtask:
Calculate and visualize the average goals scored and conceded by each team when playing at home and away. This involves grouping by 'HomeTeam' and 'AwayTeam' to get average FTHG, FTAG, and their respective conceded goals. Provide appropriate legends for the visualizations.


**Reasoning**:
To analyze each team's home performance, I will group the DataFrame by 'HomeTeam' and calculate the mean of 'FTHG' (goals scored at home) and 'FTAG' (goals conceded at home). These aggregated values will then be renamed for clarity and stored in a new DataFrame called `home_performance`.



```python
home_performance = df.groupby('HomeTeam')[['FTHG', 'FTAG']].mean().reset_index()
home_performance.rename(columns={'FTHG': 'AvgGoalsScoredHome', 'FTAG': 'AvgGoalsConcededHome'}, inplace=True)
print("Home Performance:")
print(home_performance.head())
```

**Reasoning**:
To analyze each team's away performance, I will group the DataFrame by 'AwayTeam' and calculate the mean of 'FTAG' (goals scored away) and 'FTHG' (goals conceded away). These aggregated values will then be renamed for clarity and stored in a new DataFrame called `away_performance`.



```python
away_performance = df.groupby('AwayTeam')[['FTAG', 'FTHG']].mean().reset_index()
away_performance.rename(columns={'FTAG': 'AvgGoalsScoredAway', 'FTHG': 'AvgGoalsConcededAway'}, inplace=True)
print("Away Performance:")
print(away_performance.head())
```

**Reasoning**:
Now that both home and away performance DataFrames are created, I will merge them based on the team names ('HomeTeam' and 'AwayTeam' columns) to get a complete view of each team's average goal statistics. This will involve using the `merge` function in pandas.



```python
team_performance = pd.merge(home_performance, away_performance, left_on='HomeTeam', right_on='AwayTeam', how='outer')
team_performance.rename(columns={'HomeTeam': 'Team'}, inplace=True)
team_performance.drop(columns=['AwayTeam'], inplace=True)

print("Comprehensive Team Performance:")
print(team_performance.head())
```

**Reasoning**:
To visualize the home and away goal performance, I will create a figure with two subplots. The first subplot will display a grouped bar chart for average goals scored and conceded at home for each team. The second subplot will show the same for average goals scored and conceded away. Both plots will have appropriate labels, titles, and legends for clarity.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort team_performance by total goals for better visualization
team_performance['TotalAvgGoals'] = team_performance['AvgGoalsScoredHome'] + team_performance['AvgGoalsConcededHome'] + \
                                    team_performance['AvgGoalsScoredAway'] + team_performance['AvgGoalsConcededAway']
team_performance_sorted = team_performance.sort_values(by='TotalAvgGoals', ascending=False).drop(columns=['TotalAvgGoals'])

plt.figure(figsize=(18, 10))

# Subplot 1: Home Performance
plt.subplot(1, 2, 1)
team_performance_melted_home = team_performance_sorted.melt(id_vars=['Team'],
                                                            value_vars=['AvgGoalsScoredHome', 'AvgGoalsConcededHome'],
                                                            var_name='Metric', value_name='Average Goals')
sns.barplot(x='Team', y='Average Goals', hue='Metric', data=team_performance_melted_home, palette={'AvgGoalsScoredHome': 'skyblue', 'AvgGoalsConcededHome': 'salmon'})
plt.title('Average Home Goals Scored vs. Conceded per Team')
plt.xlabel('Team')
plt.ylabel('Average Goals')
plt.xticks(rotation=90)
plt.legend(title='Goal Type', labels=['Scored Home', 'Conceded Home'])

# Subplot 2: Away Performance
plt.subplot(1, 2, 2)
team_performance_melted_away = team_performance_sorted.melt(id_vars=['Team'],
                                                            value_vars=['AvgGoalsScoredAway', 'AvgGoalsConcededAway'],
                                                            var_name='Metric', value_name='Average Goals')
sns.barplot(x='Team', y='Average Goals', hue='Metric', data=team_performance_melted_away, palette={'AvgGoalsScoredAway': 'lightgreen', 'AvgGoalsConcededAway': 'indianred'})
plt.title('Average Away Goals Scored vs. Conceded per Team')
plt.xlabel('Team')
plt.ylabel('Average Goals')
plt.xticks(rotation=90)
plt.legend(title='Goal Type', labels=['Scored Away', 'Conceded Away'])

plt.tight_layout()
plt.show()
```
